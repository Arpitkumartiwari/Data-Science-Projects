```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
df = pd.read_csv('worldometer_snapshots_April18_to_May18.csv')
df
```

## FOR USA ONLY 

```python
country_name = 'USA'
df1 = df.loc[df['Country'] == country_name, :].reset_index(drop=True)
df1
```

### FOR DATE (2020-05-18)

```python
df2= datetime.strptime('18/05/2020', '%d/%m/%Y')

df2 = df.loc[df['Date'] == df2.strftime('%Y-%m-%d'), :].reset_index(drop=True)
df2
```

### FOR LAST DATE - (2020-05-18) 

```python
df3 = datetime.strptime('18/05/2020', '%d/%m/%Y')

df3 = df.loc[df['Date'] == df3.strftime('%Y-%m-%d'), :].reset_index(drop=True)
df3
```

### PERCENTAGE OF DEATH RATE SHOULD BE (TOTAL DEATH/TOTAL CASES)*100

```python
df3['Case Fatality Ratio'] = df3['Total Deaths'] / df3['Total Cases']
plt.figure(figsize=(12,8))
plt.hist(100 * np.array(df3['Case Fatality Ratio']), bins=np.arange(35))
plt.xlabel('Death Rate (%)', fontsize=16)
plt.ylabel('Number of Countries', fontsize=16)
plt.title('Histogram of Death Rates for various Countries', fontsize=18)
plt.show()
```

## COUNTRIES WHOE WERE AT HIGH RISK

   LETS TAKE A THRESHOLD AS 1000 CASES

```python
min_number_of_cases = 1000
df4 = df3.loc[df3['Total Cases'] > min_number_of_cases,:]
plt.figure(figsize=(12,8))
plt.hist(100 * np.array(df4['Case Fatality Ratio']), bins=np.arange(20))
plt.xlabel('Death Rate (%)', fontsize=16)
plt.ylabel('Number of Countries', fontsize=16)
plt.title('Histogram of Death Rates for various Countries', fontsize=18)
plt.show()
```
